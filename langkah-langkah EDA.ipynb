- Cek nilai duplikat dan nilai unik

- Tampilkan atau visualisasikan jumlah nilai unik
    - menampilkan diagram bar nilai unik dan juga menampilkan jumlahnya dari tiap kolom dari data tersebut.

```python
# Visualisasi jumlah nilai unik di setiap kolom
check_unique.plot(kind='bar')

plt.title('Jumlah Nilai Unik per Kolom')
plt.ylabel('Jumlah Nilai Unik')

for index, value in enumerate(check_unique):
    plt.text(index, value, str(value), ha='center', va='bottom')
    
plt.show()
```

- Boxplot 
    - Memvisualisasikan Gross atau penjualan tiket dari film. Dengan ini, kita bisa mengetahui perkiraan penjualan tiket dari semua film ada di kisaran berapa dan juga mengetahui pencilan atau outlier penjualan tiket dari data tersebut.

```python
import seaborn as sns

# Membuat box plot untuk kolom 'Gross'
sns.boxplot(x='Gross', data=data)
plt.title('Box Plot of Gross')
plt.show()
```

- Korelasi
    - untuk melihat hubungan linear antara variabel-variabel numerik di dalam dataset.

```python
data_numerical = data.select_dtypes(include=['float64','int64'])

matriks_korelasi = data_numerical.corr()

plt.figure(figsize=(10, 8))  # Menyesuaikan ukuran plot agar lebih jelas
sns.heatmap(matriks_korelasi, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
